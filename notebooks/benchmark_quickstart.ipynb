# Memory Agent Eval Kit: 5-Minute Benchmark Quickstart

This notebook shows the fastest path from installation to a benchmark report using the built-in `SimpleMemoryAgent`. Replace the adapter with your own Mem0, LangGraph, or custom memory agent once the workflow is clear.

## 1. Install locally

From the repository root:

```bash
python -m venv .venv
. .venv/bin/activate
pip install -e .[dev]
```

````python
from pathlib import Path

from memory_agent_eval_kit.adapters import SimpleMemoryAgent
from memory_agent_eval_kit.benchmarks import BenchmarkRunner

agent = SimpleMemoryAgent()
run = BenchmarkRunner(agent).run(report_dir=Path("../reports/notebook_quickstart"), seed=42)
run.metrics.overall_score, run.metrics.total_scenarios
````

## 2. Inspect metrics

The default suite covers recall, contradiction handling, temporal memory, forgetting, privacy, robustness, long-horizon continuity, and multi-agent memory behavior.

````python
run.metrics.to_dict()
````

## 3. Compare agents

Run multiple adapters and build a side-by-side dashboard.

````python
from memory_agent_eval_kit.reports import write_comparison_dashboard

write_comparison_dashboard(
    {"SimpleMemoryAgent": Path("../reports/notebook_quickstart/results.json")},
    Path("../reports/notebook_quickstart"),
)
````

## 4. Export results for submission

The exchange format lets third-party agents submit comparable benchmark reports.

````python
from memory_agent_eval_kit.submissions import export_benchmark_results

export_benchmark_results(
    Path("../reports/notebook_quickstart/results.json"),
    Path("../reports/notebook_quickstart/exchange_package.json"),
    agent_name="SimpleMemoryAgent",
    benchmark_version="v0.6.0",
)
````
